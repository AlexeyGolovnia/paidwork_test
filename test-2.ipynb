```python
text = open('pizza.txt').read().lower()
```

```python
import re

text_data = re.split(r'(?<=[.!?])\s+', text)
text_data
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')

words = tokenizer.tokenize(text)
```

```python
words_set = set(words)
words_set = [{j:i} for i,j in enumerate(words_set)]
words_set
```

```python
words_set = {j:i for i,j in enumerate(words)}
words_set
```

```python
a = []
for i in text_data:  
    t = re.findall(r'(?i)([a-z]+)', i)
    try:
        a.append([words_set[i] for i in t])
    except:
        ()
    # print(t1)
```
